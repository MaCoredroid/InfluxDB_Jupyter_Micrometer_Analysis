```python

from influxdb import InfluxDBClient
from influxdb import DataFrameClient

import pandas as pd
```

```python
cli = DataFrameClient(host='202.120.40.8', port='30759',database='pizzahub')
q = "select * from http_server_requests group by Pod"
result1 = cli.query(q)
end=None
start=None
```

```python
import matplotlib.pyplot as plt
ax = None
lengedlist=[]
for onpodresult in result1:
    data=result1[onpodresult]
    data=data[data['mean']!=0]
    end=data.index[-1]
    start=data.index[0]
    ax=data.plot(y='mean',legend=onpodresult[1][0][1] ,ax=ax,figsize=(17,10))
    lengedlist.insert(len(lengedlist),onpodresult[1][0][1])
    ax.legend(lengedlist)
print(ax)
```

```python
import matplotlib.pyplot as plt
ax = None
lengedlist=[]
for onpodresult in result1:
    data=result1[onpodresult]
    data=data[data['count']!=0]
    data['diff']=data['count']/data.index.to_series().diff().dt.seconds
    ax=data.plot(y='diff',legend=onpodresult[1][0][1] ,ax=ax,figsize=(17,10))
    lengedlist.insert(len(lengedlist),onpodresult[1][0][1])
    ax.legend(lengedlist)
print(ax)
```

```python
cli = DataFrameClient(host='202.120.40.8', port='30759',database='pizzahub')
q = "select * from http_server_requests_percentile group by Pod"
result2 = cli.query(q)
```

```python
import matplotlib.pyplot as plt
ax = None
lengedlist=[]
for onpodresult in result2:
    data=result2[onpodresult]
    data=data[data['phi']=='0.95']
    data=data[data['value']!=0]
    ax=data.plot(y='value',legend=onpodresult[1][0][1] ,ax=ax,figsize=(17,10))
    lengedlist.insert(len(lengedlist),onpodresult[1][0][1])
    ax.legend(lengedlist)
print(ax)
```
